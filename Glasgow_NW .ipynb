# Glasgow GP Boundaries
In this notebook, we combine GP Boundary information from NHS Greater Glasgow and Clyde with post sector information from the National Records of Scotland to produce a map of GP boundaries of NW Glasgow. We create a shape file which can then be loaded in an ESRI application.

```python
import geopandas as gpd
import matplotlib.pyplot as plt
import pandas as pd
import shapely
from shapely.ops import unary_union
from shapely.geometry import mapping, Polygon
```

## Glasgow GP Postcodes
We first read in GP information from Glasgow.

```python
glasgowdf = pd.read_excel('./Glasgow NW Data for PHS.xls',header=None)
```

We then examine the first five rows of the data.

```python
glasgowdf.head(5)
```

Notice that the first four rows are not relevant to us. We remove them and rename our columns accordingly. Set headers correctly, remove NaN column and drop first three rows that are irrelevant. 

```python
glasgowdf.columns = glasgowdf.iloc[3] #change headers
glasgowdf = glasgowdf.dropna(axis=1, how='all') # drop the NaN column
glasgow = glasgowdf.drop(glasgowdf.index[0:4]) # drop first 3 rows
```

Double check the first few rows again. They now make more sense.

```python
glasgow.head(3)
```

We rename the postcode sector column to match the postcode sector data from NRS later on.

```python
glasgow = glasgow.rename(columns = {'pc sect (final)': 'Sector'})
```

```python
glasgow.head(3)
```

Check the dimension of the data. We expect our product to have the same number of rows (582) later.

```python
glasgow.shape
```

## Scotland Postcode Sectors Shapefile
Now we read in the postcode sector shape file from NRS.

```python
gdf = gpd.read_file('~/Desktop/GP_Boundaries/Sector_21_1/Sector_21_1.shp')
```

Check the first few items of the file.

```python
gdf.head(5)
```

Check the dimension.

```python
gdf.shape
```

As a sanity check,  we plot out the data. This is indeed a map of Scotland.

```python
gdf.plot("Shape_Area",legend=True)
```

## Merging according to postcode sector
We are only interested in the area around Glasgow City. Therefore, we now merge the two data frames and get only the postcode sectors for Glasgow that are present in the Glasgow GP information.

```python
merged = pd.merge(glasgow,gdf)
```

```python
merged.head(3)
```

Double check the dimension. This should match with the dimension of the original Glasgow data.

```python
merged.shape
```

Sort the data according to the 'CHI' numbers, which present the GP practices.

```python
#gp_merged = merged.groupby('CHI')
gp_merged = merged.sort_values(by=['CHI'])
```

```python
gp_merged.head(10)
```

Plot a few items as a sanity check.

```python
TEST = gp_merged.head(30)
test = gpd.GeoDataFrame(TEST, crs="EPSG:4326")
geomerged = gpd.GeoDataFrame(gp_merged, crs="EPSG:4326")
test.plot("CHI", legend=True)
```

```python
geomerged.plot("CHI",facecolors='none',cmap=plt.cm.jet)#,legend=True,legend_kwds={'loc':'upper right'})
plt.savefig('test_glasgow.pdf', bbox_inches='tight')
```

## Merging postcode sectors
We are interested in the GP boundaries instead of individual postcode sectors. Therefore, we dissolve the boundaries and merge the sectors.

```python
gp_geomerged = geomerged.dissolve(by='CHI')
```

```python
gp_geomerged.head(5)
```

```python
gp_geomerged.plot('Practice',facecolor='none',cmap=plt.cm.gist_ncar,legend=True,legend_kwds={'bbox_to_anchor':(2.5, 1)})
plt.savefig('dissolved.pdf', bbox_inches='tight')
```

```python
gp_geomerged.to_file('./glasgow.shp')
```
